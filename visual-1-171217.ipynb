```python
import matplotlib.pyplot as plt
import random
```

```python
# Define sample size.
size = 100
```

```python
# Method for creating a list.
def createList(sampleSize):
    randomList = []
    for i in range(sampleSize):
        randomValue = random.random()
        randomList.append(randomValue)
    return randomList
```

```python
# Return a list.
values = createList(size)
```

```python
maxindex = values.index(max(values))
minindex = values.index(min(values))

plt.figure(figsize=(20,10))
plt.xlim(-5,len(values)+5)
plt.ylim(-0.1,max(values)+0.1)
plt.annotate('Highest value', xy=(maxindex, max(values)), arrowprops=dict(facecolor='black', shrink=1))
plt.title('Random chart')
plt.xlabel('Index')
plt.ylabel('Value')
plt.xticks([maxindex,minindex])
plt.yticks([max(values),min(values)])
plt.grid(color='r', linewidth=0.1)
plt.plot(values, alpha=0.8)
plt.legend(['Wave of randomness'])
plt.show()
```
